```python
fname = "f4.vectors"
import numpy as np
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
with open(fname, "r") as f:
    ar = np.array([i.split() for i in f.readlines()], np.int32)
```

```python
M = np.array([
        [0, 2, 2, 0],
        [0, 0, 2, -2],
        [0, 0, 0, 2],
        [1, -1, -1, -1]
    ], np.int32)
```

```python
standart = ar.dot(M)
```

```python
X = PCA(2).fit_transform(standart)
```

```python
plt.plot(X[:,0], X[:,1], 'r.')
```
